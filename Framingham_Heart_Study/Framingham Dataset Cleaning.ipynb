```python
import pandas as pd 
```

```python
# Read data 
df = pd.read_csv("./data/framingham.csv")
```

```python
df.head()
```

```python
# Rename the 'male' to 'sex' for better understanding  
df.rename(columns={'male': 'sex'}, inplace=True)
```

```python
df.head()
```

```python
# 0=female and 1=male(it is not a good presentaton for summarizing data)
# Let's replace '0' with female and '1' with male 
df['sex'].replace({0: 'female', 1:'male'}, inplace=True)
```

```python
df.head()
```

```python
# Now replace '0' with non-smoker and '1' with smoker 
df['currentSmoker'].replace({0: 'nonsmoker', 1: 'smoker'}, inplace=True)
```

```python
df.head()
```

```python
# replace the level 
df['education'].replace({1: 'High School', 2: 'High School or GED', 3: 'College or Vocational', 4:'college'}, inplace=True)
```

```python
df.head()
```

```python
# 0 = Not on Blood Pressure medications; 1 = Is on Blood Pressure medications
# Replace the value 
df['BPMeds'].replace({0: 'No', 1: 'Yes'}, inplace=True)
```

```python
df.head()
```

```python
# Let's categorise BMI
def bmi_cat(bmi):
    if bmi < 18.5:
        return 'underweight'
    elif bmi > 18.5 and bmi < 24.5:
        return 'normal'
    elif bmi > 25 and bmi < 29.5:
        return 'overweight'
    else:
        return 'obese'
# Apply bmiCat function
# Function apply to DataFrame 
df['bmiCat'] = df['BMI'].apply(bmi_cat)
```

```python
df.head()
```

```python
# 0 = No; 1 = Yes
# Replace the value 
df['diabetes'].replace({0: 'No', 1: 'Yes'}, inplace=True)
```

```python
df.head()
```

```python
# 0 = No; 1 = Yes
# Replace the value 
df['prevalentHyp'].replace({0: 'No', 1: 'Yes'}, inplace=True)
```

```python
df.head()
```

```python
# 0 = No; 1 = Yes
# Replace the value 
df['prevalentStroke'].replace({0: 'No', 1: 'Yes'}, inplace=True)
```
